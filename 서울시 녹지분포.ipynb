```python
import json
import folium
import pandas as pd

jpath = './서울구별행정구역.json'
seoul_geo = json.load(open(jpath, encoding = 'utf-8')) # ;서울동별 행정구역
df = pd.read_csv('./서울구별녹지현황.csv', header = 0)
df
```

```python
g_map = folium.Map(location = [37.5642135 , 127.0016985],
                  tiles = 'openstreetmap', zoom_start = 11.2); g_map
fmap = folium.Choropleth(geo_data = seoul_geo,
                         data = df, 
                         columns = ['자치구', '면적'],
                         key_on = 'feature.properties.name', fill_color='YlGn',
                         fill_oacity = 1.5, line_0opacity = 0.3,
                         legend_name = '녹지면적',
                         highlight = True
                         ).add_to(g_map)

folium.LayerControl().add_to(g_map)

fmap.geojson.zoom_on_click = False
fmap.geojson.add_child(
    folium.features.GeoJsonTooltip(['name'],labels = False))

g_map.save('서울시녹지지구.html'); g_map
```
